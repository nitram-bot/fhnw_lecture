History:
- simple linear regression
- ANOVA
- logistic regression: Walker and Duncan 1967
- THAID: 1963?, 1977?
- CART: Breiman et al., 1984
- C4.5 Quinlan, J. R. 1993
- ID3 Quinlan, J. R. 1986

- natural cubic splines (piecwise cubic splines)
- MARS multivariate adaptive regression splines: Friedman 1991
- Vapnik et al., 1992: SVM
- random forests: 1995, 1998
- lasso: 1986, Tibshirani 1996, 
- ridge regression: Hoerl and Kennard, 1970
- boosting
- AdaBoost (adaptive boosting): Freund and Shapire 1996
- MART (multiple additive regression trees): Friedman 1999
- XGBoost: Chen et al. 2014 (community project in DMLC - distributed (deep) machine learning community)
- LightGBM: Guolin Ke 2016 (Microsoft Research)
- CatBoost: Andrey Gulin, 2017 (Yandex)
- first competition with CNNs the [dogs vs. cats](https://www.kaggle.com/c/dogs-vs-cats/leaderboard)
- Sermanet and LeCun [Traffic sign recognition with multi-scale Convolutional Networks](https://ieeexplore.ieee.org/document/6033589)
- Schmidhuber, et al.
- Hebb 1949
